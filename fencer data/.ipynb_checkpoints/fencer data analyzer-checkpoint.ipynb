# **Analysis of Fencer's SDN Data**

### Importing required packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Importing the data

```python
main_filepath = "C:/Users/TooFastDan/Documents/MD_PhD Application/Python Analysis/fencer data"
cycle_2021 = pd.read_excel(main_filepath + "/fencer_cycle_info.xlsx", sheet_name="Cycle_2020-2021", usecols="A:E")
cycle_2021
```

### Filtering the Data for graphing later

```python
pd.unique(cycle_2021["action"])
```

```python
inclusion = ['rejected groups', 'available', 'at least 1 MD/PhD acceptance']
cycle_2021 = cycle_2021[cycle_2021["action"].isin(inclusion)]
display(cycle_2021.head())
display(cycle_2021.shape)
```

```python
d0 = cycle_2021[cycle_2021["date"] == "2020-11-01"]
cycle_2021_gb = cycle_2021.groupby(["date", "action"])
cycle_2021['n'].subtract(d0['n'])
```

### Plotting some graphs

```python
plt.figure(figsize=(10, 7))
sns.lineplot(data=cycle_2021, x="date", y="n", hue="action", style="action", markers=True, dashes=False)
#plt.legend(bbox_to_anchor=(1.01, 1), borderaxespad=0) #setting the legend outside the graph
plt.title("MD/PhD Application Numbers for 2020-2021")
plt.tight_layout()
```
