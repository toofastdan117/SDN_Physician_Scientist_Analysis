```python
import re
import pandas as pd
from pandas.api.types import CategoricalDtype
import numpy as np
import math
import string
from datetime import datetime
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
import plotly.express as px
```

```python
def text_parser(filepath):
    """
    Function for parsing through a .txt file of schools with users + dates.
    Provide a filepath for each .txt file
    """
    school_list = []
    file = open(filepath)
    for line in file:
        school_list.append(line)
    return(school_list)
```

```python
def datefinder(school_list):
    """
    Function for finding the school names, complete dates, and interview invite dates.
    Provide the parsed school list as a list object.
    Returns a dataframe of school names with complete dates and interview invite dates in separate columns
    as lists of dates for each school.
    """
    schools = []
    completes = []
    interviews = []
    for school in school_list:
        
        # For finding the complete program name at the start of each line, ending with ':'
        s = re.findall(pattern='^(.*?):', string=school)
        for program in s:
            schools.append(program)
        
        # For finding all instances of complete dates starting with 'C: ' and ending with ','
        c = re.findall(pattern='C: (.*?),', string=school)
        completes.append(c)
        
        # For finding all instances of interview invite dates starting with 'II: ' and ending with ')'
        i = re.findall(pattern='II: (.*?)\)', string=school)
        interviews.append(i)
        
    # Making a dateframe with schools, complete dates, and Interview Invite Dates as the columns
    ii = pd.DataFrame({'Schools': schools, 'Complete_Dates': completes, 'Interview_Invite_Dates': interviews})
    return(ii)
```

```python
def datetimer(date_list, start_year):
    """
    A function for converting a list of strings into a list of datetime objects.
    Provide a list of dates in the form of a pandas dataframe column and the starting year of the cycle.
    Use function in a list comprehension to generate a new column of processed datetimes
    """
    dates = []
    for date in date_list:
        try:
            start = "/" + str(start_year)
            end = "/" + str(start_year + 1)
            newdate = date.split("/")
            if int(newdate[0]) < 4:
                d = date+end
                d2 = datetime.strptime(d, '%m/%d/%Y')   
                dates.append(d2)
            else:
                d3 = date+start
                d4 = datetime.strptime(d3, '%m/%d/%Y')
                dates.append(d4)
        except:
            pass
    return(dates)
```

### **Calculating the total SDN users compared to the total applicant pool**

```python
def unique_users_counter(school_list):
    """
    For making a dataframe for each SDN user and counting each unique user per cycle
    """
    schools = []
    other_users = []
    first_users = []
    for school in school_list:
        # For finding the complete program name at the start of each line, ending with ':'
        s = re.findall(pattern='^(.*?):', string=school)
        for program in s:
            schools.append(program)
        # For finding a list of users, excluding the first user
        users = re.findall(pattern='\)(.*?)\(', string=school)
        other_users.append(users)
        # For finding the first user for each school
        user1 = re.findall(pattern=':(.*?)\(', string=school)[0]
        first_users.append(user1)
    first_users = [[user1] for user1 in first_users]  # converting to a list of lists for consistency
    
    # Combining 2 lists by interating through both of them with zip
    complete_list = []
    for f, o in zip(first_users, other_users):
        combined = f + o
        complete_list.append(combined)
    
    # Making a dataframe of a list of users for each school and exploding the dataframe for each user
    sdners = pd.DataFrame({'Schools': schools, 'Users': complete_list})
    sdners = sdners.explode('Users').reset_index(drop=True)
    
    # Processing all users to remove spaces and punctuation
    processed_users = []
    for user in sdners['Users']:
        user = user.replace(" ", "") # removing spaces
        user = user.translate(str.maketrans("", "", string.punctuation)) #removing all punctuation
        processed_users.append(user)
    sdners['Processed_Users'] = processed_users
    
    # Finding the total amount of unique SDN users per cycle 
    unique_users = len(sdners['Processed_Users'].value_counts())
    
    return(sdners, unique_users)
```

```python
users_2019, uniques_2019 = unique_users_counter(school_list=school_list_2019)
users_2020, uniques_2020 = unique_users_counter(school_list=school_list_2020)
users_2021, uniques_2021 = unique_users_counter(school_list=school_list_2021)
print("Unique Users 2019: ", uniques_2019, "\nUnique Users 2020: ", uniques_2020,
     "\nUnique Users 2021: ", uniques_2021)
```

### **Calculating the days between complete date and interview invite date**

```python
def dater_timelapse(school_list, cycle):
    """
    write a description for this long-ass function
    """
    schools = []
    complete_ii = []
    for school in school_list:
        # For finding the complete program name at the start of each line, ending with ':'
        s = re.findall(pattern='^(.*?):', string=school)
        for program in s:
            schools.append(program)
        # For finding the complete and interview invite dates for each SDN user
        ci = re.findall(pattern='\((.*?)\)', string=school)
        complete_ii.append(ci)
    testdf = pd.DataFrame({'Schools': schools, 'Completes_Interviews': complete_ii})
    testdf['Cycle'] = cycle 
    testdf = testdf.explode('Completes_Interviews').reset_index(drop=True)
    
    completes = []
    interviews = []
    for ci in testdf['Completes_Interviews']:
        try:
            com = re.findall(pattern='C:(.+?),', string=ci)[0]
            com = com.replace(" ", "")
            com = com + "/" + str(2020)
            com = datetime.strptime(com, '%m/%d/%Y')
        except:
            com = np.nan
        completes.append(com)
    
        try:
            ii = re.findall(pattern='II:(.+)', string=ci)[0]
            ii = ii.replace(" ", "")
            ii_split = ii.split("/")
            if int(ii_split[0]) > 4:
                ii = ii + "/" + str(2020)
            else:
                ii = ii + "/" + str(2021)
            ii = datetime.strptime(ii, '%m/%d/%Y')   
        except:
            ii = np.nan
        interviews.append(ii)
    
    testdf['Complete_Date'] = completes
    testdf['Interview_Date'] = interviews
    testdf = testdf.dropna() # dropping any rows which contain an NA value
    testdf['timelapse'] = (testdf['Interview_Date'] - testdf['Complete_Date']).dt.days #timelapse
    
    return(testdf)
```

```python
ii_2019 = dater_timelapse(school_list=school_list_2019, cycle='2018-2019')
ii_2020 = dater_timelapse(school_list=school_list_2020, cycle='2019-2020')
ii_2021 = dater_timelapse(school_list=school_list_2021, cycle='2020-2021')

results = pd.concat([ii_2019, ii_2020, ii_2021], axis=0).reset_index(drop=True)
results
```

# **Graphing Interview Invites per School**

```python
def school_ii_plotter(df, title):
    """For plotting interview invites for each school.
    Provide a dataframe (exploded) and title as a string."""
    %matplotlib inline
    sns.set_style("ticks")
    plt.figure(figsize=(32, 12))
    plt.xticks(rotation=90)
    plot = sns.stripplot(data=df, y='Processed_Interview_Invites', x='Schools', hue='cycle')
    plot.set_xlabel("MD/PhD Programs", fontsize=16)
    plot.set_ylabel("Date of Interview Invitation", fontsize=16)
    plot.set_title(title, fontsize=24)
    plt.tight_layout()
    return(plot)
```

```python
def ii_concat(df_list):
    dfs = []
    for df in df_list:
        df1 = df.iloc[: , [0, 4, 5]]
        df2 = df1.explode('Processed_Interview_Invites').reset_index(drop=True)
        dfs.append(df2)
    results = pd.concat(dfs, axis=0).reset_index(drop=True)
    return(results)
```

```python
results = ii_concat(df_list=[ii_2019, ii_2020, ii_2021])
results
```

```python
%matplotlib
school_iis = school_ii_plotter(df=results, title="Interview Invite Dates (SDN)")
#school_iis.set_yticklabels(['August', 'September', 'October', 'November', 'December', 'January', 'February'])
#plt.savefig(all_iis.png', dpi=300)
```
