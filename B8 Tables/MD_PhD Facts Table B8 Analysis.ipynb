# **MD/PhD School Stats**

```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_excel('2020_FACTS_Table_B-8.xlsx', header=8)
df = df.iloc[0:156, ]
column_list = ["State", "Medical School", "Applications", "% Applicants In State",
               "% Applicants Out of State", "Applicant % Men", "Applicant % Women", "Matriculants",
               "% Matriculants In State", "% Matriculants Out of State", "Matriculant % Men",
               "Matriculant % Women"]
df.columns = column_list
df.set_index("Medical School", inplace=True)
applicant_filt = df["Matriculants"] > 3
df = df[applicant_filt]
df['Applicant:Matriculant Ratio'] = (df['Matriculants']/df['Applications'])*100
df['Applicant Male:Female Ratio'] = df['Applicant % Men']/df['Applicant % Women']
df['Matriculant Male:Female Ratio'] = df['Matriculant % Men']/df['Matriculant % Women']
df = df[['State', 'Applications', 'Matriculants', 'Applicant:Matriculant Ratio', 
         '% Applicants Out of State', '% Matriculants Out of State', 'Applicant Male:Female Ratio',
         'Matriculant Male:Female Ratio']]
df
```

```python
School_List = ["U Washington", 'Harvard', 'Stanford', 'UCLA-Geffen', 'UC San Diego', 'North Carolina',
              'Michigan', 'Duke', 'Vanderbilt', 'UC San Francisco', 'Baylor', 'Yale', 
               'Pennsylvania-Perelman', 'Northwestern-Feinberg', 'UT Southwestern', 'Oregon', 'Utah',
              'Wisconsin', 'Pittsburgh', 'Case Western Reserve', 'UC Irvine', 
               'Renaissance Stony Brook', 'Colorado']
schools = df.loc[School_List]
```
